### Classification using KNN

Let's set up our workspace and use the **Wine dataset** from `scikit-learn`. This dataset contains 178 wine samples with 13 chemical features, used to classify wines into different classes based on their origin.

The **response variable** is `class`, which indicates the type of wine. We'll use all of the chemical features to predict this response variable.

```python
# Import standard libraries
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import recall_score, precision_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import load_wine

# Load the Wine dataset
wine_data = load_wine()

# Convert to DataFrame
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# Bind the 'class' (wine target) to the DataFrame
wine_df['class'] = wine_data.target

# Display the DataFrame
wine_df
```

#### **Question 2:** 
#### Standardization and data-splitting

Next, we must preform 'pre-processing' or 'data munging', to prepare our data for classification/prediction. For KNN, there are three essential steps. A first essential step is to 'standardize' the predictor variables. We can achieve this using the scaler method, provided as follows:

```python
# Select predictors (excluding the last column)
predictors = wine_df.iloc[:, :-1]

# Standardize the predictors
scaler = StandardScaler()
predictors_standardized = pd.DataFrame(scaler.fit_transform(predictors), columns=predictors.columns)

# Display the head of the standardized predictors
print(predictors_standardized.head())
```

(i) Why is it important to standardize the predictor variables?

> Your answer here...

The KNN algorithm we are learning uses Euclidean distance to map the relation between predictors and a resulting dependent variable (y). Arithmetically euclidean distance is calculated adding both C1 (x1) & C (x2), so if they are not standardized predictors with larger scales will dominate the distance calculation, making them seem more important than they truly are. Likewise, smaller scales can be subestimated. Standardization takes that problem of the equation and let us add predictor distances without a problem.

(ii) Why did we elect not to standard our response variable `Class`?

> Your answer here...

Because it serves no purpose, the variable "class" being the dependent variable (y) is not used to calculate distances per se. It is used to infer the class of a new data point depending on the resulting average of the K nearest neighbours' class. Since the calculation to infer is an average of the same dependent variable but for different observations it becomes tautological.

(iii) A second essential step is to set a random seed. Do so below (Hint: use the random.seed function). Why is setting a seed important? Is the particular seed value important? Why or why not?

> Your answer here...

It is crucial for reproducibility. It ensures all random numbers are maintained to the same values se we can have consistent replicable results. This turns to be useful when debugging, comparing and validating results of a model. (iv) A third essential step is to split our standardized data into separate training and testing sets. We will split into 75% training and 25% testing. The provided code randomly partitions our data, and creates linked training sets for the predictors and response variables. 

Extend the code to create a non-overlapping test set for the predictors and response variables.

```python
# Select predictors (excluding the last column)
dependent = ["class"]
independent = wine_df.columns.difference(dependent)
```

```python

# Standardize the predictors
scaler1 = StandardScaler()
wine_df_copy[independent] = scaler1.fit_transform(wine_df_copy[independent])
independent_standardized = wine_df_copy
```

```python

# Display the head of the standardized predictors
print(independent_standardized.head())
```

```python
# set a seed for reproducibility
np.random.seed(123)

# split the data into a training and testing set. hint: use train_test_split !
wine_train, wine_test = train_test_split(
    independent_standardized, train_size=0.75, shuffle=True, stratify= independent_standardized['class']
)
```

#### **Question 3:**
#### Model initialization and cross-validation
We are finally set to fit the KNN model. 


Perform a grid search to tune the `n_neighbors` hyperparameter using 10-fold cross-validation. Follow these steps:

1. Initialize the KNN classifier using `KNeighborsClassifier()`.
2. Define a parameter grid for `n_neighbors` ranging from 1 to 50.
3. Implement a grid search using `GridSearchCV` with 10-fold cross-validation to find the optimal number of neighbors.
4. After fitting the model on the training data, identify and return the best value for `n_neighbors` based on the grid search results.

```python
knn = KNeighborsClassifier()
parameter = {
    "n_neighbors": range(1,50)
}
wine_grid = GridSearchCV(
    estimator= knn,
    param_grid= parameter,
    cv= 10
)
```

```python
wine_grid.fit(wine_train.iloc[:,:13],wine_train.iloc[:,-1])
```

```python
accuracy_grid = pd.DataFrame(wine_grid.cv_results_)
accuracy_grid
```

```python
wine_grid.best_params_
```
